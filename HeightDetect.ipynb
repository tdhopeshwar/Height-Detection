```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('height_prediction_model.h5')
```

```python
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array
```

```python

# Set the image size based on the size you used to train the model
img_size = (150, 150)  # Adjust this if your model was trained on different dimensions

# Access the laptop camera (0 is usually the built-in webcam)
cap = cv2.VideoCapture(0)

```

```python
while True:
    # Read frame-by-frame from the webcam
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to grab frame")
        break
    
    # Preprocess the frame (resize, normalize, and convert to the right shape)
    resized_frame = cv2.resize(frame, img_size)
    frame_array = img_to_array(resized_frame) / 255.0  # Normalize pixel values
    frame_array = np.expand_dims(frame_array, axis=0)  # Add batch dimension
    
    # Predict height using the loaded model
    predicted_height = model.predict(frame_array)
    
    # Display the predicted height on the frame
    cv2.putText(frame, f"Predicted Height: {predicted_height[0][0]:.2f} inches", (10, 30), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    # Show the frame with the predicted height
    cv2.imshow('Height Prediction', frame)
    
    # Press 'q' to quit the video stream
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
